```python
from crewai import Agent, Task, Crew
```

```python
planner = Agent(
    role="コンテンツプランナー",
    goal="{topic} に関する魅力的で事実に基づいたコンテンツを計画する",
    backstory="あなたは {topic} についてのブログ記事を計画しています。"
              "読者が何かを学び、情報に基づいた判断を下せるような情報を収集します。"
              "あなたの作業は、このトピックに関する記事を書くコンテンツライターのための基盤となります。",
    allow_delegation=False,
    verbose=True
)

writer = Agent(
    role="コンテンツライター",
    goal="トピック: {topic} に関する洞察に満ちた、かつ事実に基づいた意見記事を書く",
    backstory="あなたは {topic} についての新しい意見記事を書いています。"
              "その執筆は、コンテンツプランナーの作成したアウトラインと"
              "トピックに関する関連情報に基づいています。"
              "コンテンツプランナーが提供するアウトラインの主な目的と方向性に従います。"
              "また、客観的かつ公平な洞察を提供し、それをコンテンツプランナーが提供する"
              "情報で裏付けます。"
              "意見記事内で、あなたの発言が意見である場合と"
              "客観的な事実である場合を明確に区別します。",
    allow_delegation=False,
    verbose=True
)

editor = Agent(
    role="エディター",
    goal="与えられたブログ記事を組織のライティングスタイルに合わせて編集する",
    backstory="あなたは、コンテンツライターから受け取ったブログ記事を編集するエディターです。"
              "あなたの目標は、その記事がジャーナリズムのベストプラクティスに従い、"
              "意見や主張を提示する際にバランスの取れた視点を提供することを確認することです。"
              "また、可能であれば、重大な論争を引き起こすトピックや意見を避けるようにします。",
    allow_delegation=False,
    verbose=True
)
```

```python
plan = Task(
    description=(
        "1. {topic} に関する最新のトレンド、主要な関係者、注目すべきニュースを優先的に取り上げる。\n"
        "2. ターゲットオーディエンスを特定し、その興味や課題を考慮する。\n"
        "3. 導入部、主要なポイント、および行動喚起を含む詳細なコンテンツのアウトラインを作成する。\n"
        "4. SEOキーワードや関連データ、リソースを含める。"
    ),
    expected_output="アウトライン、オーディエンス分析、SEOキーワード、およびリソースを含む包括的なコンテンツ計画書。",
    agent=planner,
)

write = Task(
    description=(
        "1. コンテンツプランを活用して、{topic} に関する魅力的なブログ記事を書く。\n"
        "2. SEOキーワードを自然に取り入れる。\n"
        "3. セクションや小見出しを魅力的に命名する。\n"
        "4. 記事の構成は、読者を引きつける導入部、洞察に満ちた本文、"
        "および要約的な結論を含むようにする。\n"
        "5. 文法的な誤りを校正し、ブランドの声に合った内容に調整する。\n"
    ),
    expected_output="公開準備が整った、Markdown形式の2～3段落のセクションで構成されたブログ記事。",
    agent=writer,
)

edit = Task(
    description=(
        "指定されたブログ記事を校正し、文法的な誤りを修正し、ブランドの声に合致しているか確認する。"
    ),
    expected_output="公開準備が整ったMarkdown形式の2～3段落で構成されたブログ記事。",
    agent=editor
)
```

```python
crew = Crew(
    agents=[planner, writer, editor],
    tasks=[plan, write, edit],
    verbose=True
)
```

```python
result = crew.kickoff(inputs={"topic": "人工知能"})
```

```python
from IPython.display import Markdown

Markdown(result.raw)
```
